```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data = data.drop("Name", axis = 'columns')
```

```python
data = data.drop("Fare", axis = 'columns')
```

```python
data = data.drop("Cabin", axis = 'columns')
```

```python
data = data.drop("PassengerId", axis = 'columns')
```

```python
data = data.drop("Ticket", axis = 'columns')
```

```python
data = data.drop("Parch", axis = 'columns')
```

```python
data = data.drop("SibSp", axis = 'columns')
```

```python
lb = LabelEncoder()
data['Sex'] = lb.fit_transform(data['Sex'])
data['Embarked'] = lb.fit_transform(data['Embarked'])
```

```python
data = data.dropna()
```

```python
for column in data['Age']:
    data['Age'] = (data['Age'] -
                           data['Age'].mean()) / data['Age'].std()
```

```python
data.head()
```

```python
data.info()
```

```python
x = data.drop(["Survived"], axis= 1)
y = data["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(4,)),
    tf.keras.layers.Dense(64, activation='relu'), #Hidden
    tf.keras.layers.Dense(1, activation='sigmoid') #Output
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(x_train, y_train, epochs=50, batch_size=16, validation_split=0.4)
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy*100} %")
```
